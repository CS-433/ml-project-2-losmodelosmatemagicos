```python
import pickle
import numpy as np
import pandas as pd
```

# Data Read

```python
with open('../../data/ml4science_data.pkl', 'rb') as fp:
    data = pickle.load(fp)
```

```python
for i in range(len(data['sequences'])):
    for j in range(len(data['sequences'][i]['sequence'])):
        non_zero_count = np.count_nonzero(data['sequences'][i]['sequence'][j])
        if non_zero_count != 2:
            print(f"Vector at student {i}, sequence{j} has {non_zero_count} nonzero values")
            non_zero_positions = np.nonzero(data['sequences'][i]['sequence'][j])[0]
            print(f"Positions of non-zero values: {non_zero_positions}")
```

# Data Transform

## One hot encode - no time indication

```python
# Check
for idx in data['sequences']:
    # print(data['sequences'][idx].keys())
    # print(data['sequences'][idx]['sequence'])
    temp = [[int(cell > 0) for cell in timestep] for timestep in data['sequences'][idx]['sequence']]
    assert len(temp) == len(data['sequences'][idx]['sequence'])
    for i in range(len(temp)):
        print(data['sequences'][idx]['sequence'][i], temp[i])
    break
```

```python
# Sequence
for idx in data['sequences']:
    data['sequences'][idx]['1hot-sequence'] = [[int(cell > 0) for cell in timestep] for timestep in data['sequences'][idx]['sequence']]
    assert len(data['sequences'][idx]['1hot-sequence']) == len(data['sequences'][idx]['sequence'])
```

```python
with open('../../data/ml4science_data.pkl', 'wb') as fp:
    pickle.dump(data, fp)
```

## No break

```python
# Sequence
for idx in data['sequences']:
    data['sequences'][idx]['nobreak'] = [timestep for timestep in data['sequences'][idx]['sequence'] if timestep[-1] == 0]
    data['sequences'][idx]['nobreak'].append(data['sequences'][idx]['sequence'][-1][0:4] + [0 for i in range(5)] + [1])
```

```python
# Sequence
for idx in data['sequences']:
    data['sequences'][idx]['nobreak-1hot-sequence'] = [[int(cell > 0) for cell in timestep] for timestep in data['sequences'][idx]['nobreak']]
    assert len(data['sequences'][idx]['nobreak-1hot-sequence']) == len(data['sequences'][idx]['nobreak'])
```

```python
with open('../../data/ml4science_data.pkl', 'wb') as fp:
    pickle.dump(data, fp)
```

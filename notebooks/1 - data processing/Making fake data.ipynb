```python
with open('./ml4science_data.pkl', 'rb') as fp:
    new_dic = pickle.load(fp)
```

```python
new_dic.keys()
```

```python
fake_data = {
    'sequences': {},
    'index': {},
    'available_demographics': [d for d in new_dic['available_demographics']]
}
```

```python
fake_data['index']['learner_idx'] = {'aaa{}'.format(v): v for k, v in new_dic['index']['learner_idx'].items()}
fake_data['index']['idx_learner'] = {k: 'aaa' for k, v in new_dic['index']['idx_learner'].items()}
```

```python
for i in range(50):
    length = len(new_dic['sequences'][0]['sequence'])
    fake_data['sequences'][i] = {
        'learner_id': fake_data['index']['idx_learner'][i],
        'sequence': new_dic['sequences'][i]['sequence'][:length],
        'begin': new_dic['sequences'][i]['begin'][:length],
        'end': new_dic['sequences'][i]['end'][:length],
        'gender': np.random.choice([1, 2, 3, 4]),
        'year': np.random.choice(['1st', '2nd', '3rd']),
        'language': np.random.choice(['Deutsch', 'Fran√ßais']),
        'label': np.random.choice([0, 1])
        
    }
```

```python
for i in range(len(fake_data['sequences'][0]['begin'])):
    print(
        'begin {} - end {}: {}'.format
        (fake_data['sequences'][0]['begin'][i],
        fake_data['sequences'][0]['end'][i],
        fake_data['sequences'][0]['sequence'][i])
    )
```

```python
with open('./ml4science_data_fake.pkl', 'wb') as fp:
    pickle.dump(fake_data, fp)
```
